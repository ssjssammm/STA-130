# 1

The theoretical simple regression model is a linear relationship between a dependent and independent variable. The predictor variable (x) is the independent variable, the outcome variable (y) is the dependent variable, the slope (B1) is the change in y with an increase in x, the intercept (Bo) is the value of y when x=0 (the starting point), and the error term (E) is the independent variability of y. Random deviations from the y values predicted by the regression line is the error term.

Summary: 

In simple linear regression, the model explains the relationship between a predictor (independent variable) and an outcome (dependent variable). The predicted outcome values are determined by the intercept (the starting point on the y-axis) and the slope (the rate of change in the outcome for each one-unit increase in the predictor).

The error term represents the random differences between the actual observed values and the predicted values from the regression line. These errors are assumed to be normally distributed, centered around zero, and consistent in spread. For any fixed predictor value, the observed outcomes follow a normal distribution around the predicted line due to these random deviations.

```python
import numpy as np
import matplotlib.pyplot as plt

# Set the random seed for reproducibility
np.random.seed(0)

# Define theoretical parameters for the linear model
B0 = 2  # Intercept (starting value when x = 0)
B1 = 3  # Slope (change in y for each unit increase in x)
n = 100  # Number of observations
error_std = 5  # Standard deviation of the error term

# Generate independent variable (predictor) x
x = np.linspace(0, 10, n)

# Generate random error term E with a normal distribution
E = np.random.normal(0, error_std, n)

# Compute dependent variable y using the theoretical model: y = B0 + B1*x + E
y = B0 + B1 * x + E

# Compute the "true" regression line (without error term)
y_true = B0 + B1 * x

# Plot the theoretical model and the data with random error
plt.plot(x, y_true, color="red", label="Theoretical Regression Line (y = B0 + B1*x)")
plt.scatter(x, y, color="blue", label="Data with Error Term (y = B0 + B1*x + E)", alpha=0.6)
plt.xlabel("Independent Variable (x)")
plt.ylabel("Dependent Variable (y)")
plt.title("Theoretical Linear Regression Model with Error Term")
plt.legend()
plt.show()
```

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import statsmodels.formula.api as smf

# Set the random seed for reproducibility
np.random.seed(0)

# Define theoretical parameters for the linear model
B0 = 2  # Intercept (starting value when x = 0)
B1 = 3  # Slope (change in y for each unit increase in x)
n = 100  # Number of observations
error_std = 5  # Standard deviation of the error term

# Generate independent variable (predictor) x
x = np.linspace(0, 10, n)

# Generate random error term E with a normal distribution
E = np.random.normal(0, error_std, n)

# Compute dependent variable y using the theoretical model: y = B0 + B1*x + E
y = B0 + B1 * x + E

# Create a DataFrame with x and y values
data = pd.DataFrame({'x': x, 'y': y})

# Fit the linear regression model using statsmodels
model = smf.ols('y ~ x', data=data).fit()

# Predicted values based on the fitted model
y_fitted = model.predict(data['x'])

# Compute the "true" regression line (without error term)
y_true = B0 + B1 * x

# Plot the theoretical model, the data, and the fitted model
plt.plot(x, y_true, color="red", label="Theoretical Regression Line (y = B0 + B1*x)")
plt.scatter(x, y, color="blue", label="Data with Error Term (y = B0 + B1*x + E)", alpha=0.6)
plt.plot(x, y_fitted, color="green", linestyle="--", label="Fitted Regression Line")
plt.xlabel("Independent Variable (x)")
plt.ylabel("Dependent Variable (y)")
plt.title("Theoretical vs Fitted Linear Regression Model")
plt.legend()
plt.show()
```

```python
#1

import numpy as np
import pandas as pd
import statsmodels.formula.api as smf
import scipy.stats as stats
import plotly.express as px

np.random.seed(42)

# Bootstrapping: Fit Simple Linear Regression on Bootstrapped Samples
n = 160
long_wait_limit = 71
long_wait = old_faithful.waiting > long_wait_limit

# Collect slope coefficients from bootstrapped samples
bootstrapped_slope_coefficients = []

# Run bootstrapping (let's do 1000 resamples for a reasonable distribution)
for _ in range(1000):
    sample = old_faithful[long_wait].sample(n=n, replace=True)
    model = smf.ols('duration ~ waiting', data=sample).fit()
    bootstrapped_slope_coefficients.append(model.params['waiting'])

# Visualize bootstrapped sampling distribution of slope coefficients
fig = px.histogram(bootstrapped_slope_coefficients, nbins=50, 
                   title="Bootstrapped Sampling Distribution of Slope Coefficients")
fig.show(renderer="png")

# Calculate 95% confidence interval from bootstrapped samples
bootstrap_conf_interval = np.quantile(bootstrapped_slope_coefficients, [0.025, 0.975])
print(f"Bootstrapped 95% confidence interval for the slope coefficient: {bootstrap_conf_interval}")
```

```python
# Simulating data under the null hypothesis (no linear relationship)
old_faithful_simulation = old_faithful[long_wait].copy()
old_faithful_simulation['duration'] = 1.65 + 0 * old_faithful_simulation['waiting'] + \
                                      stats.norm(loc=0, scale=0.37).rvs(size=n)
np.random.seed(42)

# Collect slope coefficients from simulated samples
simulated_slope_coefficients = []

# Run 1000 simulations (same as bootstrapping for comparison)
for _ in range(1000):
    sample_sim = old_faithful_simulation.sample(n=n, replace=True)
    model_sim = smf.ols('duration ~ waiting', data=sample_sim).fit()
    simulated_slope_coefficients.append(model_sim.params.iloc[1])  # Access slope coefficient by position

# Visualize simulated sampling distribution of slope coefficients
fig = px.histogram(simulated_slope_coefficients, nbins=50, 
                   title="Simulated Sampling Distribution of Slope Coefficients under Null Hypothesis")
fig.show(renderer="png")

# Simulated p-value
observed_slope = smf.ols('duration ~ waiting', data=old_faithful[long_wait]).fit().params[1]
simulated_p_value = (np.abs(simulated_slope_coefficients) >= np.abs(observed_slope)).mean()
print(f"Simulated p-value: {simulated_p_value}")
```

```python
#a)

import numpy as np
import pandas as pd

# Assuming 'old_faithful' DataFrame has columns 'duration' and 'kind'
# Observed statistic: the difference in means between short and long groups
obs_diff = old_faithful.groupby('kind')['duration'].mean().iloc[::-1].diff().values[1]

# Number of permutations
n_permutations = 10000
permutation_diffs = []

# Shuffling simulation
for _ in range(n_permutations):
    shuffled_data = old_faithful.assign(kind_shuffled=old_faithful['kind'].sample(n=len(old_faithful), replace=False).values)
    perm_diff = shuffled_data.groupby('kind_shuffled')['duration'].mean().iloc[::-1].diff().values[1]
    permutation_diffs.append(perm_diff)

# Compute p-value as the proportion of permuted differences greater than or equal to the observed difference
p_value = np.mean(np.abs(permutation_diffs) >= np.abs(obs_diff))
print(f'Permutation Test p-value: {p_value}')

```

```python
#b) 

# Number of bootstrap resamples
n_bootstrap = 10000
bootstrap_diffs = []

# Bootstrap simulation
for _ in range(n_bootstrap):
    bootstrapped_data = old_faithful.groupby('kind').apply(lambda x: x.sample(n=len(x), replace=True)).reset_index(drop=True)
    bootstrap_diff = bootstrapped_data.groupby('kind')['duration'].mean().iloc[::-1].diff().values[1]
    bootstrap_diffs.append(bootstrap_diff)

# 95% Confidence Interval
ci_lower, ci_upper = np.percentile(bootstrap_diffs, [2.5, 97.5])
print(f'95% Bootstrap Confidence Interval: ({ci_lower}, {ci_upper})')
```

a) Permutation Test: Kind labels are suffled; difference in means is calculated for each dataset. obs_diff (observed differences) compared to distribution of grouped differences. P-value is proportion of more extreme differences that the observed difference.


b) Bootstrap Confidence Interval: kind groups sampled with replacement, for each resample, the mean differences are calculated. The process repeats for a bigger amount of resamples and a 95% confidence interval is constructed usinf the biitstrap differences.
